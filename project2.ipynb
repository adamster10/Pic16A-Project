```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
train_X = basketball[["W","ADJOE","ADJDE"]]
train_y = basketball["BARTHAG"]
lr.fit(train_X, train_y)
train_score = lr.score(train_X, train_y)
print('train score:', train_score)
```

```python
print('Intercept: ', lr.intercept_)  
print('Coefficients array: ', lr.coef_) 
```

```python
train_score = lr.score(train_X, train_y) 
test_score = lr.score(test_X, test_y)

print('test score:', test_score)
print(train_score)
```

```python
trainedIndex = lr.predict(train_X)
newIndex = trainedIndex + abs(trainedIndex[-1])*2 # creats array with newly created nonzero and positive index

basketball["newIndex"] = newIndex # add array to team
basketball = basketball.sort_values(by = ["newIndex"], ascending=False) # sort basketball by newIndex
```
